<font size="4">
The code is a function which will take 
    
    1) category (string) - (Make sure it's in lowercase)
    2) number of pages (int)
    
The function will return 5 lists in the following order
    
    1) Heading
    2) Summary
    3) links
    4) Date
    5) Full article

<font size="4">
  
    
    Based on their site, the only categories we can use are 
    
    1) india
    2) business
    3) world
    
 

```python
import urllib.request, sys, time
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
from tqdm import tqdm 


def scraping(category,pages):
    Heading_list = []
    Summary_list = []
    urls_list    = []
    Date_list    = []
    data_list    = []
    for page_ in tqdm(range(pages)):
        news_url = "https://www.indiatoday.in/{}?page={}".format(category,page_)

        page = requests.get(news_url)
        soup = BeautifulSoup(page.content, "html.parser")

        Article = soup.find_all("div", {"class" : "catagory-listing"})
        for j in Article:
            if j.find("div",attrs = {'class':'liveblogicon'}) == None:
                if j.find("div",attrs={'class':'pic video-icon'}) == None:
                    Heading = j.find("div",attrs={'class':'detail'}).find("h2").text.strip()
                    Summary = j.find("div",attrs={'class':'detail'}).find("p").text


                    urls = list(set([j.find("div",attrs={'class':'detail'}).find("a")["href"]]))

                    article_url =  "https://www.indiatoday.in"+urls[0]
                    article_page = requests.get(article_url)
                    article_soup = BeautifulSoup(article_page.content,"html.parser")

                    Date = article_soup.find("div",attrs={'class':'profile-detail'}).find("meta",attrs={"itemprop":"datePublished"}).get('content')
                    Date = Date[:10]

                    article_content = article_soup.find("div",{"class":"story-right"})

                    actual_content = article_content.find_all("p")
                    data =  ''.join(map(str, list(actual_content)))

                    Heading_list.append(Heading)
                    Summary_list.append(Summary)
                    urls_list.append(urls)
                    Date_list.append(Date)
                    data_list.append(data)
                    #Stuff we need to store - 
                    '''
                    1) Heading
                    2) Summary
                    3) urls
                    4) Date
                    5) data
                    '''
    return Heading_list,Summary_list,urls_list,Date_list,data_list



```
