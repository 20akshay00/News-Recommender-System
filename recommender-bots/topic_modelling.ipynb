```python
corpus = pd.read_sql_query("SELECT * FROM articles_processed ", cnx)
corpus.content
```

TFIDF vectorization

```python
vectorizer=TfidfVectorizer(max_features=1000)
vectors = vectorizer.fit_transform(corpus.content)
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
df_vecs = pd.DataFrame(dense, columns=feature_names)
df_vecs
```

LDA

```python
lda = LatentDirichletAllocation(n_components=15,learning_method='online', max_iter=20, random_state=100)
lda_matrix = lda.fit_transform(df_vecs)
lda_topics = lda.fit(df_vecs)
```

```python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print ("Topic" , topic_idx, " " .join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))
no_top_words = 10
display_topics(lda, feature_names, no_top_words)
```

LSA

```python
sv_dec_lsa = TruncatedSVD(n_components = 15)
lsa_mat = sv_dec_lsa.fit_transform(vectors)
# Sigma = sv_dec_lsa.singular_values_ 
# V_transpose = sv_dec_lsa.components_.T
terms = vectorizer.get_feature_names()
```

```python
display_topics(sv_dec_lsa, terms, no_top_words)
```
