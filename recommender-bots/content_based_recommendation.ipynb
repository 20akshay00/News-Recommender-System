```python
corpus = pd.read_sql_query("SELECT * FROM articles_processed ", cnx)
corpus.content
```

TFIDF vectorization

```python
vectorizer=TfidfVectorizer(max_features=1000)
vectors = vectorizer.fit_transform(corpus.content)
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
df_vecs = pd.DataFrame(dense, columns=feature_names)
df_vecs
```

LDA

```python
lda = LatentDirichletAllocation(n_components=25,learning_method='online', max_iter=20, random_state=100)
lda_matrix = lda.fit_transform(df_vecs)
lda_topics = lda.fit(df_vecs)
```

```python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print ("Topic" , topic_idx, " " .join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))
no_top_words = 10
display_topics(lda, feature_names, no_top_words)
```

LSA

```python
sv_dec_lsa = TruncatedSVD(n_components = 25,algorithm='randomized')
lsa_mat = sv_dec_lsa.fit_transform(vectors)
# Sigma = sv_dec_lsa.singular_values_ 
# V_transpose = sv_dec_lsa.components_.T
terms = vectorizer.get_feature_names()
```

```python
display_topics(sv_dec_lsa, terms, no_top_words)
```

```python
col = []
for i in range(lsa_mat.shape[1]):
    col.append("topic_{}".format(i)) 

topic_df = pd.DataFrame(lsa_mat, columns = col)

topic_df["Docs"] = corpus.content

l = ["Docs"]
for i in col:
    l.append(i)
topic_df
```

```python
# ------------------------------------------------------------
# START HERE FOR THE CONTENT BASED SIMILARITY.
#-------------------------------------------------------------


# Have  to manually enter the indices of the articles read by the user

article_indices = [1452, 3321, 554, 69, 420]

df_read = topic_df.iloc[[article_indices[0]]]

for i in range(len(article_indices)-1):
    # topic_df.iloc[[i]]
    df_read = pd.concat([df_read, topic_df.iloc[[article_indices[i+1]]]])
    
df_read.__delitem__('Docs')

articles_mean = df_read.mean(axis=0)
user_vector = articles_mean.to_frame().T
user_vector
```

```python
df_with_user = pd.concat([topic_df, user_vector], ignore_index=True)
df_with_user

# The last row is the required cosine similarity of the user vector with the vectorized documents.
```

```python
cos_sim_a = pd.DataFrame(cosine_similarity(df_with_user.iloc[:, :-1]))
# cos_sim_a.iloc[[-1]].__delitem__('Docs')
cos_sim_user = cos_sim_a.iloc[[-1]]
cos_sim_user.drop(cos_sim_user.columns[len(cos_sim_user.columns)-1], axis=1, inplace=True)
cos_sim_user

# Extracted the cosine similarity of user with docs row.
```

```python
best_n = cos_sim_user.apply(lambda s, n: pd.Series(s.nlargest(n).index), axis=1, n=10)
best_n = best_n.to_numpy()[0]
best_n

# Obtained the top 10 recommended article's indices.
```

```python
df_rec = topic_df.iloc[[best_n[0]]]

for i in range(len(best_n)-1):
    # topic_df.iloc[[i]]
    df_rec = pd.concat([df_rec, topic_df.iloc[[best_n[i+1]]]])
    
print('Recommended Articles')    
df_rec.iloc[:,-1:]
```
